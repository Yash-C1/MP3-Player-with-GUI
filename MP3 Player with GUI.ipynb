```python
#importing libraries
import tkinter as tkr
import random, string
import pygame
import os
from tkinter.filedialog import askdirectory
from tkinter import END
from tkinter import RIGHT,Y
#GUI Window
root = tkr.Tk()
root.geometry("800x900")
root.resizable(0,0)
root.title("MP3 Player")

pygame.init()
pygame.mixer.init()
heading = tkr.Label(root, text="YASH'S MP3 PLAYER\n", font = "100").pack()


os.chdir("/Users/meera/Desktop/Python mp3")
# Fetching Songs
song_list = os.listdir()
# Inserting Songs into Playlist

playing_state = False
first = 0
#print(song_list)

playlist_label = tkr.Label(root, text ='\nPLAYLIST\n', font = "50") 

play_list = tkr.Listbox(root,font="Helvetica 12 bold", bg="yellow", selectmode=tkr.SINGLE, height=20, width = 80)
#new_play_list = tkr.Listbox(root,font="Helvetica 12 bold", bg="yellow", selectmode=tkr.SINGLE, height=10, width = 80) 

for item in song_list:
    pos = 0
    play_list.insert(pos, item)
    pos += 1
    

def add(song_list,play_list):
    directory = askdirectory()
    os.chdir(directory)
    added_song_list = os.listdir()
    song_list += added_song_list
    
    play_list.delete(0,END)
    
    for item in song_list:
        pos = 0
        play_list.insert(pos, item)
        pos += 1
    
    
def play(first):
    pygame.mixer.music.load(song_list[first])
    pygame.mixer.music.play(0)
    first+=1   

def pause():
    global playing_state
    if not playing_state:
        pygame.mixer.music.pause()
        playing_state=True
    else:
        pygame.mixer.music.unpause()
        playing_state = False
        
    
#print(playing_state)

next_s = []


def next_song(next_s):
    next_s.append("1")
    pygame.mixer.music.load(song_list[len(next_s)])
    pygame.mixer.music.play(0)
    

def previous_song(next_s):
    pygame.mixer.music.load(song_list[len(next_s)-1])
    pygame.mixer.music.play(0)
    next_s.pop()

def stop_song():
    pygame.mixer.music.stop()
    
    

button1 = tkr.Button(root, height=1 , width= 80 , text="Add a song" ,command=lambda:add(song_list,play_list))
button1.pack()

button2 = tkr.Button(root, height=1 , width= 80 , text="Play song" ,command=lambda:play(first))
button2.pack()

button3 = tkr.Button(root, height=1 , width= 80 , text="Pause/Resume song" ,command=lambda:pause())
button3.pack()

button4 = tkr.Button(root, height=1 , width= 80 , text="Play next song" ,command=lambda:next_song(next_s))
button4.pack()

button5 = tkr.Button(root, height=1 , width= 80 , text="Play previous song" ,command=lambda:previous_song(next_s))
button5.pack()

button6 = tkr.Button(root, height=1 , width= 80 , text="Stop song" ,command=lambda:stop_song())
button6.pack()

playlist_label.pack()
play_list.pack()
#new_play_list.pack()

root.mainloop()
```
